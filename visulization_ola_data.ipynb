```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("bookings_validated.csv")

# Set style
sns.set(style="whitegrid")

# Initialize figure
plt.figure(figsize=(18, 10))

# Plot 1: Booking Status Distribution
plt.subplot(2, 3, 1)
sns.countplot(data=df, y='Booking_Status', hue='Booking_Status', order=df['Booking_Status'].value_counts().index, palette='Set2', legend=False)
plt.title('Booking Status Distribution')
plt.xlabel('Count')
plt.ylabel('Status')

# Plot 2: Vehicle Type Usage
plt.subplot(2, 3, 2)
sns.countplot(data=df, y='Vehicle_Type', hue='Vehicle_Type', order=df['Vehicle_Type'].value_counts().index, palette='Set3', legend=False)
plt.title('Vehicle Type Usage')
plt.xlabel('Count')
plt.ylabel('Vehicle Type')

# Plot 3: Payment Method Preferences
plt.subplot(2, 3, 3)
sns.countplot(data=df[df['Payment_Method'].notnull()], y='Payment_Method', hue='Payment_Method',
              order=df['Payment_Method'].value_counts().index, palette='coolwarm', legend=False)
plt.title('Payment Method Preferences')
plt.xlabel('Count')
plt.ylabel('Payment Method')

# Plot 4: Booking Value vs Ride Distance
plt.subplot(2, 3, 4)
sns.scatterplot(data=df, x='Ride_Distance', y='Booking_Value', hue='Vehicle_Type', alpha=0.6)
plt.title('Ride Distance vs Booking Value')
plt.xlabel('Ride Distance (km)')
plt.ylabel('Booking Value')

# Plot 5: Hourly Booking Trend
plt.subplot(2, 3, 5)
sns.countplot(data=df, x='Hour', hue='Hour', palette='viridis', legend=False)
plt.title('Hourly Booking Trend')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Bookings')

plt.tight_layout()
plt.show()
```
